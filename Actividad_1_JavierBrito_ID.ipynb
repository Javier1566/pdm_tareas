## 2. Define en python vectores y funciones
<p align="justify">
Debes generar vectores que representan intervalos de tiempo o de una variable.
Debes ingresar las funciones matemáticas y evaluarlas en los vectores generados.

$$
f(x) =  e^{-x^2} 
$$

$$
f(z) =  ysin(x) 
$$

<p align="justify">
Haz una gráfica para cada uno de las funciones evaluadas.
    
    

```python
import numpy as np
import matplotlib.pyplot as plt
```

* Se definen las funciones matematicas:

```python
def funcion1(x):## Funcion f(x) exponencial
    return np.exp(-1*(x**2))
    
def funcion2(x,y):## Funcion f(z) bidimensional
    return y*np.sin(x)
```

```python
x = np.linspace(-5,5,100)
y = funcion1(x) ## Se utiliza la variable 'y' para el primer grafico
```

### 2.1 Comandos para graficar f(x):

```python
plt.figure(figsize=(11,3))
plt.title("Grafico de f(x):")
plt.plot(x,y)
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.show()
```

* Se redefine 'y' para graficar la funcion f(z).

```python
y = np.linspace(-5,5,100)# Vector de valores entre -5 y 5 con 100 muestras
X, Y = np.meshgrid(x, y)## Convertir los vectores x e y en arreglos bidimensionales
Z = funcion2(X,Y)
```

### 2.2 Comandos para graficar f(z):

```python
figure = plt.figure()

ax = figure.add_subplot(111, projection = '3d')
ax.set_title("Funcion f(z):")
ax.plot_surface(X, Y, Z, cmap = 'Blues')
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.show()
```

## 3. Ruido

Crear una señal (senoidal, cosenoidal, etc) y demuestra la agregación de ruido.
    


### 3.1 Ruido Gaussiano
Debes agregar la señal de ruido a la que creaste previamente. Adicionalmente, debes mostrar una gráfica con el resultado.


### 3.1 Desarrollo de la actividad:

```python
def funcion_cos (x):
    return np.cos(x)
def funcion_sin (x):
    return np.sin(x)
    
def funcion3(x):
    return 0.6*np.cos(x-1.5)**2+1*np.sin(x-1.5)+0.2*np.cos((x-1.5)*2)
```

* Se definen las funciones para posteriormente graficarlas.

```python
xvector = np.linspace(0,15,1000)
f_cos = funcion_cos(xvector)
f_sin = funcion_sin(xvector)
f_funcion3 = funcion3(xvector)
```

### 3.1 Grafico de señales generadas :

```python
## comandos para graficar la funcion coseno
plt.figure(figsize=(11,3))
plt.subplot(1, 3, 1)
plt.plot(xvector, f_cos, color='red')
plt.title('Señal Coseno')
plt.xlabel('Eje de x')
plt.ylabel('Funcion Cosenoidal')

## comandos para graficar la funcion seno
plt.subplot(1, 3, 2)
plt.plot(xvector, f_sin, color = 'blue')
plt.title('Señal Seno')
plt.xlabel('Eje de x')
plt.ylabel('Funcion Senoidal')

## comandos para generar la funcion creada para la actividad 
plt.subplot(1, 3, 3)
plt.title("Señal periodica antes del ruido")
plt.plot(xvector,f_funcion3)
plt.xlabel("Eje x")
plt.ylabel("Funcion generada")


plt.tight_layout()
plt.show()
```

### 3.1 Grafico de ruido gaussiano :


```python
ruido_gaussiano=np.random.normal(0,1,1000) ##Se define el vector de ruido con
# media = 0 // desviacion estndar = 1 // numero de datos = 10000
## Comandos grafico de ruido gaussiano
plt.figure(figsize=(8,3))
plt.plot(xvector,ruido_gaussiano)
plt.title('Señal Ruido Gaussiano')
plt.xlabel('Eje de x')
plt.ylabel('V(x)')
plt.show()
```

* Se generan los ruidos a las señales generadas, se trabajara con 30% y 50% de ruido respectivamente.
  

```python
# Señal contaminada al 30%
cos_ruido_30 = 0.3*ruido_gaussiano + f_cos
sin_ruido_30 = 0.3*ruido_gaussiano + f_sin
funcion3_ruido_30 = 0.3*ruido_gaussiano + f_funcion3

# Señal contaminada al 50%
cos_ruido_50 = 0.5*ruido_gaussiano + f_cos
sin_ruido_50 = 0.5*ruido_gaussiano + f_sin
funcion3_ruido_50 = 0.5*ruido_gaussiano + f_funcion3
```

### 3.1 Grafico de señales contaminadas con 30% de ruido :

```python
## comandos para graficar la funcion coseno
plt.figure(figsize=(11,3))
plt.subplot(1, 3, 1)
plt.plot(xvector, cos_ruido_30, color='red')
plt.title('Señal Coseno con 30% de ruido')
plt.xlabel('Eje de x')
plt.ylabel('Funcion Cosenoidal')

## comandos para graficar la funcion seno
plt.subplot(1, 3, 2)
plt.plot(xvector, sin_ruido_30, color = 'blue')
plt.title('Señal Seno con 30% de ruido')
plt.xlabel('Eje de x')
plt.ylabel('Funcion Senoidal')

## comandos para graficar la funcion generada
plt.subplot(1, 3, 3)
plt.plot(xvector, funcion3_ruido_30, color = 'green')
plt.title('Señal periodica generada con 30% de ruido')
plt.xlabel('Eje de x')
plt.ylabel('Funcion generada')

plt.tight_layout()
plt.show()
```

### 3.1 Grafico de señales contaminadas con 30% de ruido :

```python
## comandos para graficar la funcion coseno
plt.figure(figsize=(11,3))
plt.subplot(1, 3, 1)
plt.plot(xvector, cos_ruido_50, color='red')
plt.title('Señal Coseno con 50% de ruido')
plt.xlabel('Eje de x')
plt.ylabel('Funcion Cosenoidal')

## comandos para graficar la funcion seno
plt.subplot(1, 3, 2)
plt.plot(xvector, sin_ruido_50, color = 'blue')
plt.title('Señal Seno con 50% de ruido')
plt.xlabel('Eje de x')
plt.ylabel('Funcion Senoidal')

## comandos para graficar la funcion generada
plt.subplot(1, 3, 3)
plt.plot(xvector, funcion3_ruido_50, color = 'green')
plt.title('Señal periodica generada con 50% de ruido')
plt.xlabel('Eje de x')
plt.ylabel('Funcion generada')

plt.tight_layout()
plt.show()
```
